```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import datetime
import os
```

```python
IR_path = '/home3/r05322021/Desktop/FB Data/hate_crime_data/IR/'
BH_path = '/home3/r05322021/Desktop/FB Data/hate_crime_data/BH/'
```

```python
IR_df = []
BH_df = []
for f in ['2015', '2016', '2017']:
    IR_name = IR_path + f + '.csv'
    BH_name = BH_path + f + '.csv'
    IR = pd.read_csv(IR_name)
    IR['year'] = int(f)
    BH = pd.read_csv(BH_name)
    BH['year'] = int(f)
    IR_df.append(IR)
    BH_df.append(BH)
```

```python
IR = pd.concat(IR_df, axis=0)
BH = pd.concat(BH_df, axis=0)
```

```python
df = pd.merge(IR, BH, on=['agency', 'year'], how='left')
```

```python
df = df[df.date >= 20150503]
df
```

```python
df['week'] = df['date'].apply(lambda x: np.floor((datetime.datetime.strptime(str(x), '%Y%m%d').date() - datetime.date(2015,5,3)).days / 7))
```

```python
state_code = pd.read_csv('/home3/r05322021/Desktop/FB Data/hate_crime_data/clean_data/hate_crime.csv', usecols=['state', 'state_name'])
state_code = state_code.drop_duplicates('state')
```

```python
df = pd.merge(df, state_code, on='state', how='inner')
```

```python
df['week'] = df['week'] + 1
```

```python
df.to_csv('/home3/r05322021/Desktop/FB Data/hate_crime_data/clean_data/hate_crime.csv', index=False)
```
